#1 : What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.


:-
1.**Artificial Intelligence (AI)**

Scope: Broadest field

What it is: Machines that can perform tasks requiring human intelligence.

Scope: Broader than ML; focuses on extracting insights from data.

What it is: Combines statistics, ML, data analysis, and visualization.

Techniques: Statistical modeling, ML algorithms, data cleaning, analytics.

Applications: Business analytics, forecasting, decision-making dashboards.


#2: Explain overfitting and underfitting in ML. How can you detect and prevent them?
Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.

:- Overfitting

Meaning: The model learns the training data too well, including noise → performs well on training data but poorly on test data.

Reason: High variance, low bias (model too complex).

Detection:

Large gap between training accuracy (high) and test accuracy (low).

Learning curves diverge.

Prevention:

Use cross-validation.

Apply regularization (L1/L2).

Simplify the model.

Use dropout (for neural networks).

Add more training data.

Underfitting

Meaning: Model is too simple and fails to learn patterns → performs poorly on both training and test data.

Reason: High bias, low variance (model too simple).

Detection:

Low training accuracy and low test accuracy.

Learning curves remain flat.

Prevention:

Use a more complex model.

Reduce regularization.

Add more features or useful data.

Bias–Variance Tradeoff

Overfitting → High variance

Underfitting → High bias
Goal: Find balance using cross-validation and tuning model complexity.

#3:How would you handle missing values in a dataset? Explain at least three methods with examples.
Hint: Consider deletion, mean/median imputation, and predictive modeling

:- 1. Deletion Methods

a) Listwise Deletion:
Remove entire rows with missing values.

Example: If one row has Age = NaN → delete that row.

Use when: Missing data is small (<5%) and random.

b) Pairwise Deletion:
Use available data for each analysis without removing whole rows.

2. Mean/Median/Mode Imputation

Replace missing values with a central value.

Mean → numeric data

Median → skewed numeric data

Mode → categorical data

Example:
Age column = [20, 25, NaN, 30]
Mean = 25
Replace NaN with 25.

3. Predictive Modeling (Advanced Imputation)

Use other features to predict the missing value using ML models.

Algorithms: KNN Imputer, Regression imputation, Random Forest Imputer.

Example:
Predict missing Salary using features like Experience, Age, Education.

Summary

Deletion → When missing data is small.

Mean/Median Imputation → Simple & fast, but reduces variability.

Predictive Modeling → More accurate but computationally expensive.

#4:What is an imbalanced dataset? Describe two techniques to handle it(theoretical + practical).
Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.

:- Imbalanced Dataset

An imbalanced dataset is when classes are not represented equally—
one class has many more samples than the other.

Example:
Fraud detection →

Legit transactions: 98%

Fraud: 2%
This causes models to predict the majority class more often.

Techniques to Handle Imbalanced Data
1. Random Oversampling / Undersampling
a) Oversampling

Increase samples of the minority class by duplicating data.

Theory:
Balances dataset by adding minority examples.

Practical Example (Python):

from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
X_res, y_res = ros.fit_resample(X, y)

b) Undersampling

Reduce samples from the majority class.

Theory:
Simplifies dataset by removing majority examples.

Practical Example:

from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler()
X_res, y_res = rus.fit_resample(X, y)

2. SMOTE (Synthetic Minority Oversampling Technique)

Theory:
Creates synthetic new minority samples using nearest neighbors instead of simple duplication.

Practical Example:

from imblearn.over_sampling import SMOTE
sm = SMOTE()
X_res, y_res = sm.fit_resample(X, y)

3. Class Weights (Model-based method)

Increase penalty for misclassifying minority class.

Theory:
Model learns to treat minority class as more important.

Practical Example (Logistic Regression):

from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(class_weight='balanced')
clf.fit(X, y)



#5: Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.
Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient
descent.


:- Feature scaling means bringing all numerical features into a similar range (like 0–1 or -1 to 1). It helps ML models treat all features equally and improves model accuracy and training speed.

Impact of Feature Scaling on ML Algorithms

Prevents large-value features from dominating
Example: Salary (0–1,000,000) vs Age (18–70)

Without scaling → salary dominates all calculations.

Improves distance-based algorithms
Algorithms that depend on distances get more accurate after scaling.

Helps gradient-based algorithms converge faster
Gradient descent moves smoothly when all features are on the same scale.

Algorithms that NEED Scaling

Algorithm Reason

KNN Uses distance → scaling mandatory K-Means Cluster formation depends on distances SVM Margin + distance calculations Logistic Regression Uses gradient descent Neural Networks Stable gradients, faster training

Algorithms that DO NOT Need Scaling

Algorithm Reason

Decision Tree Splits based on thresholds, not distances Random Forest Tree-based → scaling not required XGBoost / LightGBM Also tree-based

Example

If a dataset has:

Age: 25

Salary: 50,000

Salary is very large compared to age. Distance-based models will treat salary as more important. Scaling fixes this by making both features comparable.

Conclusion

Feature scaling is essential for distance-based and gradient-based models but unnecessary for tree-based algorithms.

#6.Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?
Hint: Consider categorical variables with ordinal vs. nominal relationships.

:- Label Encoding

What it does: Converts categories into numeric labels (0, 1, 2…).

Best for: Ordinal data — where order matters.
Examples: Low < Medium < High, Small < Medium < Large.

Issue: For nominal data, it creates a false numeric relationship.

Example:
Color → {Red: 0, Blue: 1, Green: 2}
Model may think Green > Blue > Red, which is wrong.

One-Hot Encoding

What it does: Creates separate binary columns for each category.

Best for: Nominal data — no natural order.
Examples: Gender, City, Color.

Pros: No false ordering.

Cons: Increases dimensionality (many new columns).

Example:
Color →
Red → [1,0,0]
Blue → [0,1,0]
Green → [0,0,1]

When to Prefer Which?
Encoding	Prefer When	Reason
Label Encoding	Ordinal categories	Order matters (e.g., size, rating)
One-Hot Encoding	Nominal categories	No order; avoids false numeric meaningv

#7: Google Play Store Dataset a). Analyze the relationship between app categories and ratings. Which categories have the
highest/lowest average ratings, and what could be the possible reasons?
Dataset: https://github.com/MasteriNeuron/datasets.git
(Include your Python code and output in the code box below.)


```python
import pandas as pd

# 1. Load the dataset
# Make sure the CSV file is in the same folder as your notebook
df = pd.read_csv("googleplaystore.csv")

# 2. Check basic information
print(df.head())
print(df.info())

# 3. Clean rating column
# Convert "Rating" column to numeric
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Remove rows with missing rating or category
df = df.dropna(subset=['Rating', 'Category'])

# 4. Calculate average rating by Category
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

# 5. Show result
print("Average Rating by Category:\n")
print(category_ratings)

# 6. Highest & Lowest rated categories
highest = category_ratings.idxmax()
lowest = category_ratings.idxmin()

print("\nHighest Rated Category:", highest)
print("Lowest Rated Category:", lowest)
```

#8: Titanic Dataset a) Compare the survival rates based on passenger class (Pclass). Which class had the highest survival rate, and why do you think that happened? b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and
adults (Age ≥ 18). Did children have a better chance of survival?
Dataset: https://github.com/MasteriNeuron/datasets.git
(Include your Python code and output in the code box below.)

```python
# 1. Load the dataset
df = pd.read_csv("titanic.csv")   # Make sure file name is correct

# 2. Basic info
print(df.head())
print(df.info())

# 3. Survival rate by Passenger Class
class_survival = df.groupby('Pclass')['Survived'].mean()

print("\nSurvival Rate by Passenger Class:")
print(class_survival)

highest_class = class_survival.idxmax()
print("\nClass with highest survival rate:", highest_class)


# 4. Create Age Group column (Child / Adult)
df['AgeGroup'] = df['Age'].apply(lambda x: "Child" if x < 18 else "Adult")

# 5. Survival rate by Age Group
age_survival = df.groupby('AgeGroup')['Survived'].mean()

print("\nSurvival Rate by Age Group:")
print(age_survival)
```

#9: Flight Price Prediction Dataset a) How do flight prices vary with the days left until departure? Identify any exponential price surges and recommend the best booking window. b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
consistently cheaper/premium, and why?
Dataset: https://github.com/MasteriNeuron/datasets.git
(Include your Python code and output in the code box below.)

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load dataset
df = pd.read_csv("flight_price.csv")   # change name if needed

# 2. Basic information
print(df.head())
print(df.info())

# 3. How price changes with days left until departure
price_by_days = df.groupby("days_left")["price"].mean()

print("\nAverage price by days left until departure:")
print(price_by_days.head())

# Plot price trend (optional if you want visualization)
plt.figure(figsize=(8,5))
plt.plot(price_by_days.index, price_by_days.values)
plt.xlabel("Days Left Until Departure")
plt.ylabel("Average Price")
plt.title("Price Trend vs Days Left")
plt.show()


# 4. Compare airline prices for same route (Example: Delhi to Mumbai)
route_df = df[(df["source_city"] == "Delhi") & (df["destination_city"] == "Mumbai")]

airline_price = route_df.groupby("airline")["price"].mean().sort_values()

print("\nAverage price by Airline (Delhi → Mumbai):")
print(airline_price)


# 5. Cheapest & Most Expensive airline
cheapest = airline_price.idxmin()
expensive = airline_price.idxmax()

print("\nCheapest Airline:", cheapest)
print("Most Expensive Airline:", expensive)
```

# 10: HR Analytics Dataset a). What factors most strongly correlate with employee attrition? Use visualizations to show key drivers (e.g., satisfaction, overtime, salary). b). Are employees with more projects more likely to leave?
Dataset: hr_analytics


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("HR_Analytics.csv")

print("Columns in dataset:\n", df.columns)

# 'left' = Attrition (1 = left, 0 = stayed)


# (a) FACTORS AFFECTING ATTRITION


# 1. Satisfaction level vs Attrition
plt.figure(figsize=(6,4))
sns.barplot(x='left', y='satisfaction_level', data=df)
plt.title("Satisfaction Level vs Attrition")
plt.show()

# 2. Number of Projects vs Attrition
plt.figure(figsize=(6,4))
sns.barplot(x='number_project', y='left', data=df)
plt.title("Number of Projects vs Attrition")
plt.show()

# 3. Monthly hours vs Attrition
plt.figure(figsize=(6,4))
sns.boxplot(x='left', y='average_montly_hours', data=df)
plt.title("Working Hours vs Attrition")
plt.show()

# 4. Salary vs Attrition
plt.figure(figsize=(6,4))
sns.countplot(x='salary', hue='left', data=df)
plt.title("Salary vs Attrition")
plt.show()

# 5. Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation with Attrition")
plt.show()



# (b) ARE MORE PROJECTS = MORE ATTRITION ?

print("\nAttrition rate by number of projects:\n")
print(df.groupby("number_project")["left"].mean() * 100)
```
